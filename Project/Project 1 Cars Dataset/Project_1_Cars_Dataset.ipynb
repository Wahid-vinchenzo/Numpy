```python
import pandas as pd #import pandas
import matplotlib.pyplot as plt #import matplotlib
```

```python
car =pd.read_csv("https://raw.githubusercontent.com/Wahid-vinchenzo/Numpy/refs/heads/main/PANDAS/file.csv") #import csv
```

```python
car.head() #show top 5 file of csv
```

```python
car.shape #check the element number usinng car shape (432 x 15)
```



# Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column 🙄
 *data cleaning*

```python
nul=car.isnull() #check the null value using isnull function it's show a bolean table
nul=nul.sum() # count the number of null values per columns
nul

car = car.fillna(car.mean(numeric_only=True)) #fill the null value with the mean of columns for only numeric data type
nul

```

*(Based on Value Counts)*

## Check what are the different types of Make are there in our dataset. And, what is the count (occurrence) of each Make in the data? 😜

```python
car.head(2)
```

```python
mk=car['Make'] # showing the differennt types of Make
mk
mk.value_counts() #show the values count of different kind of Make
```

*instruction (Filtering)*
## Show all the records where Origin is Asia or Europe.😎

```python
car.head(2)
```

```python
org= car['Origin'].isin(['Asia','Europe']) #show all the records where Origin is Asia or Europe.
car[org]
```

*Instruction (Removing unwanted records)*

Remove all the records (rows) where Weight is above 4000 🥰

```python
car.head(2)
```

```python
ovrw=car['Weight'] > 4000 #find the weight value are more than 4000
car=car[~ovrw] #remove the weight value are more than 4000
car
```

```python
car.shape #after removing the unwanted data the shape of dataset
```

*instruction (Applying function on a column)*

## Increase all the values of MPG City' column by 3 😁

```python
car.head(2)
```

```python
mpg_manip=car['MPG_City'].apply(lambda x:x+3) #increase all the values of MPG City' column by 3
mpg_manip
car['MPG_City']=mpg_manip #update the dataset
car
```

# ***1️⃣ Histogram of 'Horsepower' 😮***

```python

plt.figure(figsize=(6, 4)) #specified the histogram size
plt.hist(car['Horsepower'], bins=20, color='skyblue', edgecolor='black') #title
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## ***2️⃣ Scatter Plot: EngineSize vs Horsepower ✔***

```python

plt.figure(figsize=(6, 4)) #specified the diagram size
plt.scatter(car['EngineSize'], car['Horsepower'], color='green')
plt.title('Engine Size vs Horsepower')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.grid(True)
plt.show()
```

# ***3️⃣ Pie Chart of Car 'Type' 😍***

```python

type_counts = car['Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Car Type Distribution')
plt.show()
```

# ***4️⃣ Subplots: Histogram + Scatter in 1 row 😂***

```python

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Subplot 1: Histogram
axs[0].hist(car['Horsepower'], bins=20, color='purple', edgecolor='black')
axs[0].set_title('Histogram of Horsepower')
axs[0].set_xlabel('Horsepower')
axs[0].set_ylabel('Frequency')
# Subplot 2: Scatter
axs[1].scatter(car['EngineSize'], car['Horsepower'], color='orange')
axs[1].set_title('Engine Size vs Horsepower')
axs[1].set_xlabel('Engine Size')
axs[1].set_ylabel('Horsepower')

plt.tight_layout()
plt.show()
```
