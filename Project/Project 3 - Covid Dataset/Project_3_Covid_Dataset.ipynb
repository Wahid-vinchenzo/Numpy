## **✅✅code for  bar and pie charts for the top 10 countries by deaths and recovered cases from  dataset 👋**

```python
# Grouping by region and summing confirmed, deaths, and recovered
summary = cov.groupby('Region')[['Confirmed', 'Deaths', 'Recovered']].sum()

# Get top 10 regions by Deaths and Recovered
top10_deaths = summary.sort_values(by='Deaths', ascending=False).head(10)
top10_recovered = summary.sort_values(by='Recovered', ascending=False).head(10)

# Set the style
sns.set(style='whitegrid')

# Set up the plot
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Bar plot for top 10 regions by Deaths
top10_deaths['Deaths'].plot(kind='bar', ax=axs[0, 0], color='darkred')
axs[0, 0].set_title('Top 10 Regions by Deaths')
axs[0, 0].set_ylabel('Deaths')
axs[0, 0].tick_params(axis='x', rotation=45)

# Pie chart for top 10 regions by Deaths
axs[0, 1].pie(top10_deaths['Deaths'], labels=top10_deaths.index, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Death Distribution (Top 10 Regions)')

# Bar plot for top 10 regions by Recovered
top10_recovered['Recovered'].plot(kind='bar', ax=axs[1, 0], color='green')
axs[1, 0].set_title('Top 10 Regions by Recovered')
axs[1, 0].set_ylabel('Recovered')
axs[1, 0].tick_params(axis='x', rotation=45)

# Pie chart for top 10 regions by Recovered
axs[1, 1].pie(top10_recovered['Recovered'], labels=top10_recovered.index, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Recovery Distribution (Top 10 Regions)')

# Adjust layout
plt.tight_layout()
plt.show()
```
